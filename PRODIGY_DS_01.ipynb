### Population Distribution

```python
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']

total_population_top = total_population_data.sort_values(by="2001", ascending=False)

top_ten_countries = total_population_top.head(10)
print("Top ten countries of total population\n")
print(top_ten_countries[['Country Code']])
```

```python
year = ('2022', '2021', '2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2010',
       '2009', '2008', '2007', '2006', '2005', '2004', '2003', '2002', '2001')
```

```python
for i in year:
    plt.figure(figsize=(15,6))
    plt.subplot(2,2,1)
    sns.barplot(x=i, y="Country Code", data=top_ten_countries, palette="coolwarm")
    plt.title(f"Top ten countries of total population {i}")
    plt.xlabel("Total population")
    plt.ylabel("Country")
    plt.show()
```

```python
total_population_b = total_population_data.sort_values(by="2022", ascending=True)

bottom_ten_countries = total_population_b.head(10)
print("Bottom ten countries of total population\n")
print(bottom_ten_countries[['Country Code']])
```

```python
for i in year:
    plt.figure(figsize=(15,6))
    plt.subplot(2,2,1)
    sns.barplot(x=i, y="Country Code", data=bottom_ten_countries, palette="coolwarm")
    plt.title(f"Bottom ten countries of total population {i}")
    plt.xlabel("Total population")
    plt.ylabel("Country")
    plt.show()
```

### Stacked Bar Plot

```python
merge_data=pd.merge(male_population,female_population,on="Country Code",suffixes=("_male","_female"))
merge_data["Total population"] = merge_data["2022_male"] + merge_data ["2022_female"]
```

```python
merge_data.head()
```

```python
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

```python
top_10 = sorted_data.head(10)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x="Country Code",y="2022_female",data=top_10,color="red",label="female Population")
sns.barplot(x="Country Code",y="2022_male",data=top_10,bottom=top_10["2022_female"],color="green",label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
bottom_10 = sorted_data.tail(10)
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x="Country Code", y="2022_female", data=bottom_10, color="red", label="Female Population")
sns.barplot(x="Country Code", y="2022_male", data=bottom_10, bottom=bottom_10["2022_female"], color="green", label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
